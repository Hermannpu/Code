```python
def MAE (window_size,file):
     x = np.empty([11712-window_size+1 , 9 * (window_size-1)], dtype = float)
     y = np.empty([11712-window_size+1, 1], dtype = float)
     
    
     for i in range(11712-window_size+1): #滑动窗口的大小为24，取前23个数据放入x，第24个的output放入y
        x[i,:] = file[i:window_size+i-1:,].reshape(1,-1)  #将前23个数据平铺成一行向量
        y[i,0] = file[window_size+i-1,8]                  #将第24行的油温加入y
        x_train = x[: math.floor(len(x) * 0.75), :]       #前12个月作为数据集
        y_train = y[: math.floor(len(y) * 0.75), :]
        x_validation = x[math.floor(len(x) * 0.75): , :]   #后4个月作为测试集
        y_validation = y[math.floor(len(y) * 0.75): , :]
     model = LinearRegression()
     model.fit(x_train, y_train)
     y_pred = model.predict(x_validation)
     MAE = metrics.mean_absolute_error(y_validation, y_pred)
     return MAE

def MSE (window_size,file):
     x = np.empty([11712-window_size+1 , 9 * (window_size-1)], dtype = float)
     y = np.empty([11712-window_size+1, 1], dtype = float)
     
    
     for i in range(11712-window_size+1): #滑动窗口的大小为24，取前23个数据放入x，第24个的output放入y
        x[i,:] = file[i:window_size+i-1:,].reshape(1,-1)  #将前23个数据平铺成一行向量
        y[i,0] = file[window_size+i-1,8]                  #将第24行的油温加入y
        x_train = x[: math.floor(len(x) * 0.75), :]       #前12个月作为数据集
        y_train = y[: math.floor(len(y) * 0.75), :]
        x_validation = x[math.floor(len(x) * 0.75): , :]   #后4个月作为测试集
        y_validation = y[math.floor(len(y) * 0.75): , :]
     model = LinearRegression()
     model.fit(x_train, y_train)
     y_pred = model.predict(x_validation)
     MSE = metrics.mean_squared_error(y_validation, y_pred)
     return MSE
```
